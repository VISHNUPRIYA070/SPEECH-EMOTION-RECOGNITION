```python
# NOISE
def noise(data):
    noise_amp = 0.035*np.random.uniform()*np.amax(data)
    data = data + noise_amp*np.random.normal(size=data.shape[0])
    return data
# STRETCH
def stretch(data, rate=0.8):
    return librosa.effects.time_stretch(data, rate=rate)
# SHIFT
def shift(data):
    shift_range = int(np.random.uniform(low=-5, high = 5)*1000)
    return np.roll(data, shift_range)
# PITCH
def pitch(data, sampling_rate, pitch_factor=0.7):
    return librosa.effects.pitch_shift(data,sr=sampling_rate, n_steps=pitch_factor)
```

```python
def feat_ext(data):
    mfcc = np.mean(librosa.feature.mfcc(y=data, sr=sample_rate).T, axis=0)
    return mfcc

def get_feat(path):
    data, sample_rate = librosa.load(path, duration=2.5, offset=0.6)
    # normal data
    res1 = feat_ext(data)
    result = np.array(res1)
    #data with noise
    noise_data = noise(data)
    res2 = feat_ext(noise_data)
    result = np.vstack((result, res2))
    #data with stretch and pitch
    new_data = stretch(data)
    data_stretch_pitch = pitch(new_data, sample_rate)
    res3 = feat_ext(data_stretch_pitch)
    result = np.vstack((result, res3))
    return result
```

```python
X, Y = [], []
for path, emotion in zip(df['path'], df['Emotion']):
    feature = get_feat(path)
    for ele in feature:
        X.append(ele)
        Y.append(emotion)
```

```python
Features = pd.DataFrame(X)
Features['labels'] = Y
Features.to_csv('features.csv', index=False)
Features.head()
```

```python
X = Features.iloc[: ,:-1].values
Y = Features['labels'].values
```

```python
# As this is a multiclass classification problem onehotencoding our Y.
encoder = OneHotEncoder()
Y = encoder.fit_transform(np.array(Y).reshape(-1,1)).toarray()
```

```python
# splitting data
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=0, shuffle=True)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf1=KNeighborsClassifier(n_neighbors=5)
clf1.fit(x_train,y_train)
```

```python
y_pred=clf1.predict(x_test)
```

```python
print("Training set score: {:.3f}".format(clf1.score(x_train, y_train)))
print("Test set score: {:.3f}".format(clf1.score(x_test, y_test)))
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear')

# Convert y_train to 1D array before fitting
y_train_1d = np.argmax(y_train, axis=1)
# fitting x samples and y classes
clf.fit(x_train, y_train_1d) # Use the 1D target variable here

# For prediction, convert the predictions back to one-hot encoded format
y_pred = clf.predict(x_test)
y_pred_onehot = to_categorical(y_pred)

print("Training set score: {:.3f}".format(clf.score(x_train, y_train_1d)))
print("Test set score: {:.3f}".format(clf.score(x_test, np.argmax(y_test, axis=1))))
```
